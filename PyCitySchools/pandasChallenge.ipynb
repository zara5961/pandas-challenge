```python
# First, import needed libraries
import pandas as pd
import csv
import numpy as np

# Call the files through the path 
school_data_load = "Resources/schools_complete.csv"
student_data_load = "Resources/students_complete.csv"

# Read both Data files and give them the name to save as a DataFrame
school_data = pd.read_csv(school_data_load)
student_data = pd.read_csv(student_data_load)

# Merge two DataFrames base on school name (which is common between both csv), and since we want to have statistical info on students we do the left merge to keep all rows in student data.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Start Statictical Analysis
# The total number of schools
total_school = len(school_data_complete['school_name'].unique())
total_school
```

```python
# The total number of students
total_student = school_data_complete['student_name'].count()
total_student
```

```python
# And the total budget
total_budget = sum(school_data_complete['budget'].unique())
total_budget
```

```python
# start with calculating the average math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# Then calculate the average reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# Calculate the overall passing rate(math &reading: since it is matematical clulation it is easy to sum them up and divide it by two)
overall_passing_score = (average_math_score + average_reading_score)/2
overall_passing_score
```

```python
# Calculate the percentage of students with a pass math score (70 or greater)
passing_math_score = (school_data_complete[school_data_complete['math_score']>=70]['student_name'].count()/total_student)*100
passing_math_score
```

```python
# Calculate the percentage of students with a pass reading score (70 or greater)
passing_reading_score = (school_data_complete[school_data_complete['reading_score']>=70]['student_name'].count()/total_student)*100
passing_reading_score
```

```python
# Calculate the total number of students
student_count = school_data_complete['Student ID'].nunique()
student_count

#Calculate the percentage of students that pass both math & reading according their score
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a Dataframe with given name to have the above results and formatting
district = {
    'Total number of unique Schools':total_school,
    'Total Students':'{:,}'.format(total_student),
    'Total Budget':'${:,.2f}'.format(total_budget),
    'Average Math Score':average_math_score,
    'Average Reading Score':average_reading_score,
    '% Passing Math':passing_math_score,
    '% Passing Reading':passing_reading_score,
    '% Overall Passing Score':[overall_passing_score],  
}

district_summery = pd.DataFrame(district)
district_summery
```

```python
# # Create a Dataframe with given name to have the above results and formatting (School name is already set as index)
school={
    'School Type': school_type,
    'Total Students':total_student,
    'Total School Budget': total_budget,
    'Per Student Budget': t_budget_per_student,
    'Average Math Score': average_math_score,
    'Average Reading Score': average_reading_score,
    '% Passing Math': percent_passing_math,
    '% Passing Reading': percent_passing_reading,
    '% Overall Passing Rate': percent_overall_passing,
}
school_summary = pd.DataFrame(school)
# Create a copy of school summary data frame before formatting to be able to use the numeric data on original data frame later
displayed_school_summary = school_summary.copy()
# Formatting the display data frame
displayed_school_summary['Per Student Budget'] = displayed_school_summary['Per Student Budget'].map('${:,.2f}'.format)
displayed_school_summary['Total School Budget'] = displayed_school_summary['Total School Budget'].map('${:,.2f}'.format)
displayed_school_summary.index.name = None

school_summary.head()
```

```python
# bins & labels
spending_bins = [0, 585, 615, 645, 675]
labels  = ["<$585", "$585-615", "$615-645", "$645-675"]
# Create a new Data frame by locating all the desired columns
scores_spending = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate',]]
# I added a new column named Spending Ranges (Per Student) and binning based off budget per student
scores_spending['Spending Ranges (Per Student)']= pd.cut(school_summary['Per Student Budget'],spending_bins,labels=labels)


# Create a group based on the bins
scores_spending = scores_spending.groupby('Spending Ranges (Per Student)').mean()
scores_spending.head()
```

```python
# Sample bins
size_bins = [0, 1000, 2000, 5000]
labels  = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
# Create a new Data frame by locating the desired columns
scores_size = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate',]]
# I added a new columns named School Size and binning based off total students
scores_size['School Size']= pd.cut(school_summary['Total Students'],size_bins,labels=labels )
# Create a group based off of the bins
scores_size = scores_size.groupby('School Size').mean()
scores_size.head()
```
